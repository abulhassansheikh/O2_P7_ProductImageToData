```python
## Model: L/W/H -> Weight
#  Multivariable Regression

#Compile Dimentional Data to construct model
DimData = ImageRef.merge(Image2Height, on= "image_name", how="left") 
DimData = DimData[~DimData["image_name"].duplicated(keep=False)] 

DimData = DimData.merge(Image2Length, on= "image_name", how="left") 
DimData = DimData[~DimData["image_name"].duplicated(keep=False)] 

DimData = DimData.merge(Image2Width, on= "image_name", how="left") 
DimData = DimData[~DimData["image_name"].duplicated(keep=False)] 

DimData = DimData.merge(Image2Weight, on= "image_name", how="left") 
DimData = DimData[~DimData["image_name"].duplicated(keep=False)] 

#Decleare Height, Width, length, Weight as numeric columns and remove any text
DimData["height"] = pd.to_numeric(DimData["height"], errors='coerce')
DimData["length"] = pd.to_numeric(DimData["length"], errors='coerce')
DimData["width"] = pd.to_numeric(DimData["width"], errors='coerce')
DimData["weight"] = pd.to_numeric(DimData["weight"], errors='coerce')

#Remove any NaN values 
DimData = DimData[
             ~DimData["height"].isna() &
             ~DimData["length"].isna() &
             ~DimData["width"].isna() &
             ~DimData["weight"].isna()]

#Lets create volume column also
DimData["volume"] = DimData["height"] * DimData["length"] * DimData["width"] 

#Create MVR model
Dim_model = smf.ols(formula = 'weight ~ height+ length+ width + volume', data=DimData)
#Dim_model = smf.ols(formula = 'weight ~ height+ length+ width', data=DimData)

#Consider SVM Reggression, Tree regression, something that is not linear

Dim_model_Val = Dim_model.fit()
Dim_model_Val.summary()

#Since all values have a p-value <0.05, and the R-sqr value is reduced with the
#removal of any variable, I will keep  all variables in the model.
#There is a multi multicollinearity warning due to volume, but given the 
#very small volume coefficent, it can help account for extreme values in one of the
#L/W/H variables.

intercept = Dim_model_Val.params[0]
heightCoef = Dim_model_Val.params[1]
lenghtCoef = Dim_model_Val.params[2]
widthCoef = Dim_model_Val.params[3]
volumeCoef = Dim_model_Val.params[4]
Dim_model_Val.summary()
```

```python
## Model: PT -> L/W/H
#  Compute Mean and Standard Deviation 
#Compile Dimentional Data to construct model
PTRef = pd.DataFrame({"part_type":Image2PT["part_type"].dropna().unique()})

PTDimData = ImageRef.merge(Image2PT, on= "image_name", how="left") 
PTDimData = PTDimData[~PTDimData["image_name"].duplicated(keep=False)] 

PTDimData = PTDimData.merge(Image2Height, on= "image_name", how="left") 
PTDimData = PTDimData[~PTDimData["image_name"].duplicated(keep=False)] 

PTDimData = PTDimData.merge(Image2Length, on= "image_name", how="left") 
PTDimData = PTDimData[~PTDimData["image_name"].duplicated(keep=False)] 

PTDimData = PTDimData.merge(Image2Width, on= "image_name", how="left") 
PTDimData = PTDimData[~PTDimData["image_name"].duplicated(keep=False)] 

#Decleare Height, Width, length, Weight as numeric columns and remove any text
PTDimData["height"] = pd.to_numeric(PTDimData["height"], errors='coerce')
PTDimData["length"] = pd.to_numeric(PTDimData["length"], errors='coerce')
PTDimData["width"] = pd.to_numeric(PTDimData["width"], errors='coerce')

#Remove any NaN values from Part Type
PTDimData = PTDimData[~PTDimData["part_type"].isna()]
alldim = ["height", "length", "width"]

#Parameters for controling dimentional confidence interval
error = 0.1
interval = st.norm.ppf(1-(error)/2)

#Mean & STD of all three dimentions per PT
for d in range(len(alldim)):
    value = alldim[d]

    SingDimData = pd.DataFrame(PTDimData.groupby("part_type")[value].agg(["count", "mean", np.std])).reset_index()    
    SingDimData["interval"]  = interval* (SingDimData["std"] /np.sqrt(SingDimData["count"]))
    SingDimData[value+"_max"] = SingDimData["mean"] + SingDimData["interval"] 
    SingDimData[value+"_min"] = SingDimData["mean"] - SingDimData["interval"] 
    SingDimData = SingDimData[["part_type", "mean", value+"_max", value+"_min"]]
    SingDimData.columns = ["part_type", value+"_mean", value+"_max", value+"_min"]

    PTRef = PTRef.merge(SingDimData, on ="part_type", how = "outer")
```

```python
#Predict PT weight 

for p in range(len(PTRef)):
    pt_prediction = PTRef.loc[p, "part_type"]
    print(pt_prediction)
    
    PTpreDim = PTRef[PTRef["part_type"] == pt_prediction].reset_index()
    height_max = PTpreDim["height_max"][0]
    height_min = PTpreDim["height_min"][0]
    length_max = PTpreDim["length_max"][0]
    length_min = PTpreDim["length_min"][0]
    width_max = PTpreDim["width_max"][0]
    width_min = PTpreDim["width_min"][0]
    volume_max = height_max * length_max * width_max
    volume_min = height_min * length_min * width_min

    weight_min = (intercept + 
                (heightCoef * height_min) +
                (lenghtCoef * length_min) +
                (widthCoef * width_min) + 
                (volumeCoef * volume_min) )

    weight_max = (intercept + 
                (heightCoef * height_max) +
                (lenghtCoef * length_max) +
                (widthCoef * width_max) +
                (volumeCoef * volume_max) )
    actualweight = ImageRefDF[ImageRefDF["part_type"]==pt_prediction][["weight"]].reset_index()
    
    if len(actualweight)>0:
        for w in range(len(actualweight)):
            weightVal = actualweight.loc[w, "weight"]
            if (weight_min<weightVal<weight_max)==True:
                actualweight.loc[w, "Pred"]  = 1
            else:
                actualweight.loc[w, "Pred"]  = 0            
    #        elif (weightVal<weight_min)==True:
    #            actualweight.loc[w, "Pred"]  = "Under"
    #        elif (weight_max<weightVal)==True:
    #            actualweight.loc[w, "Pred"]  = "Over"

        PTRef.loc[p, "Prediction"] = actualweight["Pred"].sum()/len(actualweight["Pred"])

PTRef_Comp = PTRef[
             ~PTRef["height_max"].isna() &
             ~PTRef["height_min"].isna() &
             ~PTRef["length_max"].isna() &
             ~PTRef["length_min"].isna() &
             ~PTRef["width_max"].isna() &
             ~PTRef["width_min"].isna() ]
```
